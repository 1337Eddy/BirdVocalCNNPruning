Data Preparation

```python
path = "birdclef/"
path_soundscape = path + "train_soundscapes/"
path_shortaudio = path + "train_short_audio/"
path_soundscape_labels = path + "train_soundscape_labels.csv"
path_train_meta = path + "train_metadata.csv"

train_csv = pd.read_csv(path_train_meta)
soundscape_csv = pd.read_csv(path_soundscape_labels)

train_labels = list(train_csv.value_counts('primary_label')[:50].index)

train_batch = train_csv[train_csv['primary_label'].isin(train_labels)]
train_batch
```

```python
sr = 32000
n_mels=128
fmin = 0
fmax = None
sequenceLen = 130
batchSize = 30

class AudioDataset(Sequence):
    def __init__(self, data, sr=sr, fmin=fmin, fmax=fmax, n_mels=n_mels):
        self.data = data
        self.sr = sr
        self.fmin = fmin
        self.fmax = fmax
    def __len__(self):
        return len(self.data)
    
    def reshape(self, arr):
        output = np.reshape(arr, len(arr[0]) * len(arr))
        pad = (n_mels * sequenceLen * math.ceil(np.shape(arr)[1] / sequenceLen)) - len(output)
        output =np.pad(output, (0, pad), 'constant', constant_values=0)
        
        labels = np.zeros((math.ceil(np.shape(arr)[1] / sequenceLen), len(train_labels)))
        
        X = np.reshape(output, (math.ceil(np.shape(arr)[1] / sequenceLen), 128, 130))
        return X, labels
    
    def read_file(self, path):
        sig, sr = sf.read(path)
        
        melspec = lb.feature.melspectrogram(sig, sr=sr, n_mels=n_mels, fmax=fmax, fmin=fmin)
        melspec = lb.power_to_db(melspec).astype(np.float32)
        
        return self.reshape(melspec)
    
        
    
    def __getitem__(self, idx):
        name = list(train_batch['primary_label'])[idx]
        data, labels = self.read_file(path_shortaudio + list(train_batch['primary_label'])[idx] + "/"+ list(train_batch['filename'])[idx])
        index_item = np.where(np.array(train_labels) == name)
        for i in range(0,len(labels)):
            labels[i][index_item] = 1
        return data, labels
```

```python
data = AudioDataset(train_batch)
len(data.__getitem__(3)[0])
```

```python
model = tf.keras.Sequential([
    
    # First conv block
    tf.keras.layers.Conv2D(filters=16, kernel_size=5, activation='relu', 
                           input_shape=(None, n_mels, sequenceLen)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    
    # Second conv block
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)), 
    
    # Third conv block
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)), 
    
    # Fourth conv block
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    
    # Global pooling instead of flatten()
    tf.keras.layers.GlobalAveragePooling2D(), 
    # Dense block
    tf.keras.layers.Dense(256, activation='relu'),   
    tf.keras.layers.Dropout(0.5),  
    tf.keras.layers.Dense(256, activation='relu'),   
    tf.keras.layers.Dropout(0.5),
    

    # Classification layer
    tf.keras.layers.Dense(units=len(train_labels), activation='softmax')
])
print('MODEL HAS {} PARAMETERS.'.format(model.count_params()))

model.compile(optimizer=Adam(learning_rate=2e-3), loss='mean_squared_error', metrics=['accuracy'])
```

```python
history = model.fit(x=data[0],y=data[1], epochs = 4)
```
